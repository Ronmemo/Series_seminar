```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore') # 計算警告を非表示
```

# データの準備

```python
df = pd.read_csv('Bitcoin.csv', thousands=',')
df.head() # DataFrame先頭5行
```

```python
df = df.iloc[::-1]
df.head() # DataFrame先頭5行
```

```python
df.tail()
```

```python
bit = pd.Series(df['終値'], dtype='float')
bit.index = pd.to_datetime(df['日付け'], format = '%Y年%m月%d日')

print(bit.index.dtype)
print(len(bit))
bit.head()
```

```python
plt.figure(figsize=(12,8))
bit.plot()
```

# データの分析

```python
#分解
res = sm.tsa.seasonal_decompose(bit)

original = bit # オリジナルデータ
trend = res.trend # トレンドデータ
seasonal = res.seasonal # 季節性データ
residual = res.resid # 残差データ

plt.figure(figsize=(8, 8)) # グラフ描画枠作成、サイズ指定

# オリジナルデータのプロット
plt.subplot(411) # グラフ4行1列の1番目の位置（一番上）
plt.plot(original)
plt.ylabel('Original')

# trend データのプロット
plt.subplot(412) # グラフ4行1列の2番目の位置
plt.plot(trend)
plt.ylabel('Trend')

# seasonalデータ のプロット
plt.subplot(413) # グラフ4行1列の3番目の位置
plt.plot(seasonal)
plt.ylabel('Seasonality')

# residual データのプロット
plt.subplot(414) # グラフ4行1列の4番目の位置（一番下）
plt.plot(residual)
plt.ylabel('Residuals')

plt.tight_layout()
```

```python
kion_month_mean = bit.groupby(bit.index.month).mean()
kion_month_mean.plot(kind='bar')
```

```python
#コレログラム
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(bit.values.squeeze(), lags=80, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(bit, lags=80, ax=ax2)
```

# モデル推定

```python
# 単位根検定
adf_result = sm.tsa.stattools.adfuller(bit)
adf_result
# 棄却できない→単位根でないとはいえない
```

```python
#対数変換
log_bit = np.log(bit)

plt.figure(figsize=(12,8))
plt.plot(log_bit)
```

```python
#コレログラム
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(log_bit.values.squeeze(), lags=80, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(log_bit, lags=80, ax=ax2)
```

```python
#対数差分変換
logdiff_bit = np.log(bit) - np.log(bit.shift())
logdiff_bit = logdiff_bit.dropna()

plt.figure(figsize=(12,8))
plt.plot(logdiff_bit)
```

```python
#コレログラム
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(logdiff_bit.values.squeeze(), lags=80, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(logdiff_bit, lags=80, ax=ax2)
```

```python
# 自動ARMAパラメータ推定関数
res_selection = sm.tsa.arma_order_select_ic(log_bit, max_ar=3, max_ma=3, ic='aic', trend='nc')
res_selection
```

```python
# SRIMAモデル作成その１
sarimax = sm.tsa.SARIMAX(log_bit, 
                        order=(1, 1, 2),
                        seasonal_order=(1, 1, 2, 12),
                        enforce_stationarity = False,
                        enforce_invertibility = False
                        ).fit()
```

```python
#コレログラム
sarimax_resid = sarimax.resid 

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(sarimax_resid, lags=80, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(sarimax_resid, lags=80, ax=ax2)
```

```python
print(sarimax.aic)
```

```python
sarimax.summary()
```

```python
sarimax_pred = sarimax.predict('2015-06', '2019-06') 

plt.figure(figsize=(8, 4))

#元データ
plt.plot(log_bit, label="original")
#モデル
plt.plot(sarimax_pred, c="b", label="model-pred", alpha=0.7)
plt.legend(loc='best')
```

```python
#先1年まで予測
sarimax_pred = sarimax.predict('2019-01', '2019-12') 

plt.figure(figsize=(8, 4))

plt.plot(log_bit, label="original")
plt.plot(sarimax_pred, c="b", label="model-pred", alpha=0.7)
plt.legend(loc='best')
```

# train & predict

```python
bit_train = log_bit['2015-06':'2018-12'] # モデル作成用データ
print(bit_train.head())
print(bit_train.tail())
```

```python
bit_test = log_bit['2019-01':'2019-6'] # テスト用データ
print(bit_test.head())
print(bit_test.tail())
```

```python
# SRIMAモデル（テストデータを除いてモデル作成）
sarimax_train = sm.tsa.SARIMAX(bit_train, 
                        order=(1, 1, 2),
                        seasonal_order=(1, 1, 2, 12),
                        enforce_stationarity = False,
                        enforce_invertibility = False
                        ).fit()
```

```python
sarimax_train_pred = sarimax_train.predict('2019-01', '2019-06') # テストデータ予測

plt.figure(figsize=(12, 8))

plt.plot(log_bit['2018-01':'2019-06'], label="original")
plt.plot(sarimax_train_pred, c="b", label="model-pred", alpha=0.7)
plt.legend(loc='best')
```

```python
plt.figure(figsize=(8, 4))

plt.plot(bit_test, label="actual") # 正解
plt.plot(sarimax_train_pred, c="b", label="predict", alpha=0.7) # 予測
plt.legend(loc='best')
```

```python
predict_dy = sarimax_train.get_prediction(start ='2019-01', end='2019-06')
type(predict_dy)
```

```python
predict_dy_ci = predict_dy.conf_int(alpha=0.05) #95%信頼区間
type(predict_dy_ci)
```

```python
predict_dy_ci # lower, upper取得
```

```python
predict_dy.predicted_mean # mean取得
```

```python
plt.figure(figsize=(8, 4))

plt.plot(bit_test, label="Actual")
plt.plot(predict_dy.predicted_mean, c="b", label="model-pred", alpha=0.7)

plt.fill_between(predict_dy_ci.index, predict_dy_ci.iloc[:, 0], predict_dy_ci.iloc[:, 1], color='g', alpha=0.2)
plt.legend(loc='upper left')
```

```python
predict2_dy = sarimax_train.get_prediction(start ='2019-01', end='2019-12') # 1年分の未来予測

predict2_dy_ci = predict2_dy.conf_int()

plt.figure(figsize=(12, 8))
plt.plot(log_bit['2015-06':], label="actual")
plt.plot(predict2_dy.predicted_mean, c="b", linestyle='--', label="model-pred", alpha=0.7)

plt.fill_between(predict2_dy_ci.index, predict2_dy_ci.iloc[:, 0], predict2_dy_ci.iloc[:, 1], color='g', alpha=0.2)
plt.legend(loc='upper left')
```
